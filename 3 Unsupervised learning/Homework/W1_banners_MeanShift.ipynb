```python
import pandas as pd
import numpy as np
from sklearn.cluster import MeanShift
```

```python
data = pd.read_csv('fsq/checkins.dat', delimiter='|', skipinitialspace=True).dropna(axis=0)
```

```python
len(data)
```

```python
data.head(5)
```

```python
X = data[['latitude','longitude']]
```

```python
X
```

```python
ms = MeanShift(bandwidth=0.1)
ms.fit(X)
```

```python
labels_unique = np.unique(ms.labels_)
n_clusters = len(labels_unique)
n_clusters
```

```python
list((ms.labels_))
```

```python
mslight = MeanShift(bandwidth=0.1, min_bin_freq=15)
mslight.fit(X[:100000])
```

```python
light_labels_unique = np.unique(mslight.labels_)
len(light_labels_unique)
```

```python
len(ms.labels_)
```

```python
labels = ms.labels_
centers = ms.cluster_centers_
```

```python
from collections import Counter
```

```python
interesting_labels = [label for label in labels_unique if Counter(labels)[label]>15]
```

```python
len(interesting_labels)
```

```python
points = [centers[i] for i in range(len(centers)) if interesting_labels.count(labels[i])>0]
len(points)
```

```python
Counter(labels)
```

```python
offices = []
offices.append([33.751277, -118.188740])
offices.append([25.867736, -80.324116])
offices.append([51.503016, -0.075479])
offices.append([52.378894, 4.885084])
offices.append([39.366487, 117.036146])
offices.append([-33.868457, 151.205134])
best_points=[]
dist=[]
for point in points:
    best_dist=1000000
    for office in offices:
        d = np.sqrt((point[0]-office[0])*2+(point[1]-office[1])*2)
        if d<best_dist:
            best_dist=d
    dist.append(best_dist)

dist = np.array(dist)
idxs = np.argsort(dist)[:20]
```
