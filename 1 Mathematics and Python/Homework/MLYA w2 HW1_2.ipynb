```python
import numpy as np
import scipy as sp
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fx = lambda x: np.sin(x/5.0) * np.exp(x/10.0) + 5.0 * np.exp(-x / 2.0)
```

```python
xx = np.arange(1.0, 15.0, 0.1)
```

```python
plt.plot (xx, fx(xx))
```

```python
powr = 1
```

```python
A1 = [
    [1 ** n for n in range(0, powr + 1)],
    [15 ** n for n in range(0, powr + 1)]
]
```

```python
b1 = [fx(1.0), fx(15.0)]
```

```python
fx1 = sp.linalg.solve(A1, b1)
```

```python
fx1
```

```python
x1 = np.array([1.0, 15.0])
```

```python
plt.plot (x1, fx1)
plt.plot (xx, fx(xx))
plt.show()
```

```python
powr = 2
```

```python
A2 = [
    [1 ** n for n in range(0, powr+1)],
    [8 ** n for n in range(0, powr+1)],
    [15 ** n for n in range(0, powr+1)]
]
```

```python
b2 = [fx(1.0), fx(8.0), fx(15.0)]
```

```python
fx2 = sp.linalg.solve(A2, b2)
```

```python
fx2
```

```python
plt.plot ([1.0, 8.0, 15.0], fx2)
plt.plot (xx, fx(xx))
plt.show()
```

```python
powr = 3
```

```python
A3 = [
    [1 ** i for i in range(0, powr + 1)],
    [4 ** i for i in range(0, powr + 1)],
    [10 ** i for i in range(0, powr + 1)],
    [15 ** i for i in range(0, powr + 1)]
]

b3 = [fx(1), fx(4), fx(10), fx(15)]
```

```python
fx3 = sp.linalg.solve(A3, b3)
```

```python
fx3
```

```python
plt.plot ([1.0, 4.0, 10.0, 15.0], fx3)
plt.plot (xx, fx(xx))
plt.show()
```
